```
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns

%matplotlib inline

data_bci= pd.read_csv("J1_Test (1).csv")
data_bci.iloc[:,1: 17].head()

```

```
fig = plt.figure(figsize=(12, 9))
ax1  = fig.add_subplot(1, 1, 1)

ax1.scatter(data_bci['CH1'], data_bci['CH6'], c=['#FFE888' if sh==1 else('#FF9999' if sh==2 else 'blue') for sh in data_bci['Shape']])

ax1.set_xlabel("CH1")
ax1.set_ylabel("CH6")

ax1.set_title("Scatter plot of CH1 and CH6")

plt.show()
```

```
fig = plt.figure(figsize=(12, 9))
ax2  = fig.add_subplot(111, projection='3d')

ax2.scatter(data_bci['CH6'], data_bci['CH7'], data_bci['CH9'], alpha=0.4, s=30, c= ['#FF9999' if sh == 1 else ('#FFE888' if sh==2 else 'blue') for sh in list(data_bci['Shape'])] )

ax2.set_xlabel("CH6")
ax2.set_ylabel("CH7")
ax2.set_zlabel("CH9")

ax2.set_title("Scatter plot of CH6 and CH7 and CH9")
plt.savefig("Scatter_plot_of_CH6_CH7_CH9.png")
plt.show()
```
